```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
col_boundary_2015 = ("https://stacks.stanford.edu/file/druid:sb666tw0101/data.zip")


```

```python
# Country boundary 2015
col_boundary_2015 = ("https://stacks.stanford.edu/file/druid:sb666tw0101/data.zip")
col_boundary_gdf = gpd.read_file(col_boundary_2015)
col_boundary_gdf
```

```python
# Administrative boundaries ()
admin_url = "https://maps.princeton.edu/download/file/tufts-colombia-municipalities-11-shapefile.zip"
admin_url
```

```python
admin_gdf = gpd.read_file(admin_url)
admin_gdf
```

```python
# Print all the DEPT names
print(admin_gdf['NOM_DEPART'].to_string())
```

```python
# Extract PUTUMAYO
putumayo_boundary = admin_gdf[admin_gdf["NOM_DEPART"].isin(["PUTUMAYO"])]
putumayo_boundary
```

```python
# Petroleum exploration 2013
oil_url = ("https://maps.princeton.edu/download/file/tufts-colombia-petroleumgasexplotation-sites-13-shapefile.zip")

oil_gdf = gpd.read_file(oil_url)
oil_gdf
```

```python
fig, ax = plt.subplots(figsize=(6,10))
col_boundary_gdf.plot(ax=ax, edgecolor='grey', color='beige')
oil_gdf.plot(ax=ax,edgecolor='black', color='None')
```

```python
# Clip oil/gas to Putumayo boundary
putumayo_petrol = gpd.clip(oil_gdf, putumayo_boundary)
putumayo_petrol
```

```python
# Map of all the oil/gas sites in Putumayo (2013)
fig, ax = plt.subplots()
putumayo_boundary.plot(ax=ax, color='beige', edgecolor='black', alpha=0.5)
putumayo_petrol.plot(ax=ax, color='None', edgecolor='black')
plt.show()
```

```python
# Zoom in on PUERTO ASÍS

# Extract PUERTO ASÍS from Putumayo boundary
puerto_asis = putumayo_boundary[putumayo_boundary["NOM_MUNICI"].isin(["PUERTO ASÍS"])]
puerto_asis

# Clip oil gas to PUERTO ASÍS boundary
puerto_asis_oil_gas = gpd.clip(putumayo_petrol, puerto_asis)
puerto_asis_oil_gas

# Plot Puerto Asis boundary and clipped oil and gas
fig, ax = plt.subplots()
puerto_asis.plot(ax=ax, color='beige', edgecolor='grey')
puerto_asis_oil_gas.plot(ax=ax, color='None', edgecolor='black')
plt.show()
```
